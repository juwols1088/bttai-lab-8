First I am preparing the data for the model by normalizing the values through a standard scaler. 

```python
# 1. Create a StandardScaler object and save the result to variable scaler.
scaler = StandardScaler()
# 2. Call the scaler.fit_transform() method to fit the scaler to data_to_scale
# and tranform the data. Save the result to transformed_data.
transformed_data = scaler.fit_transform(df)
```

```python
#4. Create new DataFrame df_scaled

df_scaled = pd.DataFrame(transformed_data, columns = df.columns, index = df.index)

# Inspect df_scaled
df_scaled
```

Now the data is ready to be split and trained on the model: 
here I split the data into the features (X) and the label (y)

```python
X = df_scaled.drop(columns = 'Life Ladder', axis = 1)
y = df_scaled['Life Ladder']
```

Next I split the data into a training set of x and y and a test set of x and y. I have decided to use 0.3 which means 30% of the data is randomly assigned to the test set and 70% is assigned to the train data.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

Below, I create the model, I fit the model to the training data set and then I make a prediction using the test set and the model I just fit. 

```python
# Create the  LinearRegression model object 
model = LinearRegression()

# Fit the model to the training data 
model.fit(X_train, y_train)

#  Make predictions on the test data 
prediction = model.predict(X_test)
```

```python
#viewing the prediction
prediction
```

I run RMSE and R^2 as evaluation metrics.

```python
# The mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
# The coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

Mean square error is very low which is good since RMSE is a measure of loss and we want our model to have a very loss to show that it is accurate. It's correlation R^2 is very close to 1 which means the features are very strongly correlated in a linear relationship with the label. It seems like the model is relatively accurate at capturing the relationship between the features and the label based on these metrics. Below is more specific information about the weights and intercept of the linear model and the line of its equation.

```python
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', X.columns[i],']')
    i += 1
```

```python
#getting the weights
weights = model.coef_
```

```python
#printing weights
weights
```

```python
#plotting the weights
sns.barplot(x = X.columns, y = weights)
plt.xticks(rotation=90)
```
